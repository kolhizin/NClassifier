```python
def plot_imgregion(axes, imgregion, **kwargs):
    if type(imgregion) in (list, tuple):
        for x in imgregion:
            plot_imgregion(axes, x, **kwargs)
        return
    xs = [imgregion[0], imgregion[1], imgregion[1], imgregion[0], imgregion[0]]
    ys = [imgregion[2], imgregion[2], imgregion[3], imgregion[3], imgregion[2]]
    axes.plot(xs, ys, **kwargs)
    
def plot_point(axes, pnt, imgsize, **kwargs):
    axes.plot((0.5+0.5*pnt[0])*imgsize[0], (0.5 - 0.5*pnt[1])*imgsize[1], 'bo', **kwargs)
    
def plot_region(axes, center, direction, ortho, size, img_size, **kwargs):
    coefs = [np.array([-size[0], size[1]]), np.array([size[0], size[1]]), np.array([size[0], -size[1]]), np.array([-size[0], -size[1]]), np.array([-size[0], size[1]])]
    points = [(center + c[1] * direction + c[0] * ortho) for c in coefs]
    xs = [(0.5+0.5*p[0])*img_size[0] for p in points]
    ys = [(0.5-0.5*p[1])*img_size[1] for p in points]
    axes.plot(xs, ys, **kwargs)
```

```python
def rotate_point(point, rot_matrix, old_size, new_size):
    old_abs = point * np.array(old_size)
    new_abs = np.matmul(rot_matrix.transpose(), old_abs)
    return new_abs / np.array(new_size)

def adjust_region_for_image_rotation(region, rot_matrix, old_size, new_size):
    """
    Adjust region for rotation of image.
    :param region: region to be adjusted.
    :param rot_matrix: rotation matrix used for image rotation.
    :param old_size: old size of image
    :param new_size: new size of image
    :return: adjusted region
    """
    np_old_size = np.array(old_size)
    np_new_size = np.array(new_size)
    
    
    res_center = np.matmul(rot_matrix.transpose(), region[0] * np_old_size) / np_new_size    
    res_dir = np.matmul(rot_matrix.transpose(), region[1] * np_old_size) / np_new_size
    res_ortho = np.matmul(rot_matrix.transpose(), region[2] * np_old_size) / np_new_size
        
    tmp_dir_len = np.sqrt(np.dot(res_dir, res_dir))
    tmp_ortho_len = np.sqrt(np.dot(res_ortho, res_ortho))
    res_dir = res_dir / tmp_dir_len    
    res_ortho = res_ortho / tmp_ortho_len        
        
    return res_center, res_dir, res_ortho, region[3]*np.array([tmp_ortho_len, tmp_dir_len])
```
